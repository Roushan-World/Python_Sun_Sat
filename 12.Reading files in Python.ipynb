# 13. Reading Files in Python

 * __To read a text file for reading file in Pyhton, follow these steps:__
 * __First, Open a text file for reading by using the open() functions.__
 * __Second, tead text from the text file using the file read(), readline(), or readlines() method of the file object.__
 * __Third, close the file close() method. This frees up resources and ensures consistency across different python version.__

```python
x='C:\\Users\\Admin\\Downloads\\Chandra_Shekhar_Azad.txt'
file=open(x,'r')
content=file.read()
content
```

```python
print(content)
```

```python
# printing the path of file
print(file.name)
```

```python
# printing the mode of file

print(file.mode)
```

```python
# printing the type of file..

print(type(content))
```

```python
# verification of file closed or not?

file.closed
```

```python
file.close()
```

```python
y='C:\\Users\\Admin\\Downloads\\Chandra_Shekhar_Azad.txt'

with open(y,'r') as f:
    print(f.read(10))
    print(f.read(100))
    print(f.read(150))
```

```python
y='C:\\Users\\Admin\\Downloads\\Chandra_Shekhar_Azad.txt'

with open(y,'r') as f:
    print(f.readline())
    print(f.readline())
    print(f.readline())
    print(f.readline())
```

```python
y='C:\\Users\\Admin\\Downloads\\Chandra_Shekhar_Azad.txt'

with open(y,'r') as f:
    print(f.readlines(1))
    print(f.readlines(1))
    print(f.readlines(1))
```
